# 4. Fill in all the missing values present in all the columns in the dataset

```python
data.info()
```

There are missing values int he Age,Cabin and Embarked group

```python
plt.hist(data['Age'])
plt.show()
```

It is a left skewed group so we use the median instead of mean

```python
data['Age']=data['Age'].fillna(data['Age'].median())
```

```python
data['Cabin']=data['Cabin'].fillna(data['Cabin'].mode()[0])
```

```python
data['Embarked']=data['Embarked'].fillna(data['Embarked'].mode()[0])
```

# 5. Check and handle outliers in at least 3 columns in the dataset

```python
plt.boxplot(data['Age'])
plt.show
```

```python
#caluculating the quartiles in the data.
Q1=np.percentile(data['Age'],25,interpolation='midpoint')
Q2=np.percentile(data['Age'],50,interpolation='midpoint')
Q3=np.percentile(data['Age'],75,interpolation='midpoint')
print('Q1 = ',Q1,'\nQ2 = ',round(Q2,2),'\nQ3 = ',Q3)
```

```python
#Calculating Inter Quartile Range
IQR=Q3 - Q1
print('IQR =',IQR)
```

```python
#calculating the upper and lower limits of outliers
low_lim=Q1 -(1.5*IQR)
upper_lim=Q3+(1.5*IQR)
print('lower limit=',low_lim,'upper limit=',upper_lim)
```

```python
#creating list of outliers 
outlier=[]
for x in data['Age']:
    if (x>upper_lim or x <low_lim):
        outlier.append(x)
outlier
```

```python
#Finding the index values of outliers
ind1 =data['Age']<low_lim
ind2=data['Age']>upper_lim
data.loc[ind1|ind2].index
```

```python
ind3=data.loc[ind1|ind2].index
```

```python
#removing outliers
data.drop(ind3,inplace=True)
```
